**Data Visualisation**

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 4 , 8, 12 , 16]
y = [3, 5 , 7, 9 ,8]

plt.figure(figsize = (8,4))
plt.plot(x , y , marker = '*' , c='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = [1, 4 , 8, 12 , 16]
y = [3, 5 , 7, 9 ,8]

plt.figure(figsize = (8,4))
plt.scatter(x , y , marker = '*' , c='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = ['A' , 'B' , 'C' , 'D' , 'E']
y = [5 , 7 , 12 , 3 , 6]

plt.bar(x,y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/bipulshahi/Dataset/refs/heads/main/imdb_1000.csv')
df.head(3)
```

```python
#scatter plot between star_rating & duration
plt.scatter(df.duration , df.star_rating)
plt.xlabel('Duration')
plt.ylabel('Star Rating')
plt.show()
```

```python
df.genre.value_counts()
```

```python
print(df.genre.value_counts().index)
print()
print(df.genre.value_counts().values)
```

```python
plt.figure(figsize = (8,4))
plt.bar(df.genre.value_counts().index , df.genre.value_counts().values)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (8,7))
plt.barh(df.genre.value_counts().index , df.genre.value_counts().values)
plt.show()
```
