```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/bipulshahi/Dataset/refs/heads/main/imdb_1000.csv')
df.head(3)
```

```python
#scatter plot between star_rating & duration
plt.scatter(df.duration , df.star_rating)
plt.xlabel('Duration')
plt.ylabel('Star Rating')
plt.show()
```

```python
df.genre.value_counts()
```

```python
print(df.genre.value_counts().index)
print()
print(df.genre.value_counts().values)
```

```python
plt.figure(figsize = (8,4))
plt.bar(df.genre.value_counts().index , df.genre.value_counts().values)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (8,7))
plt.barh(df.genre.value_counts().index , df.genre.value_counts().values , color = 'green')
plt.show()
```

```python
genre_count = df.genre.value_counts().reset_index()
genre_count
```

* if count > 100 => 'Green'
* if count > 50 => 'Yellow'
* else => 'Red'

```python
def generate_color(value):
    if value > 100:
        return 'green'
    elif value > 50:
        return 'yellow'
    else:
        return 'red'
```

```python
bar_color = genre_count['count'].apply(generate_color)
bar_color
```

```python
plt.figure(figsize = (8,7))
plt.barh(genre_count['genre'] , genre_count['count'], color = bar_color)
plt.show()
```
